# DCG: Algoritmo whoosh standard 

```python
from vanilla import ricerca as w
from benchmark import bench as b

benchmark_standard=None
rt=w.Search(condition="g", limit=10 , package="vanilla")
benchmark_standard=b.Mark(rt , cond= "g" )
benchmark_standard.target_obj.stampa_bench=benchmark_standard.dummy_print
for i in benchmark_standard.query_set:
    _ = benchmark_standard.target_obj.search(i[0], advanced=i[1])
benchmark_standard.target_obj.DCG=[]
algo_whoosh=benchmark_standard.target_obj.get_DCG_counter()
print(algo_whoosh)
```
